```python
column_names = Anshux.columns #JadesCam.columns

# Create an empty DataFrame to store the results
Fluxes = pd.DataFrame()

# Iterate over the column names
for column_name in column_names:
    # Check if the column is a flux column (e.g., 'F105W_flux')
    if column_name.endswith('_flux'):
        error_column = column_name[:-5] + '_err'
        if error_column in column_names:
            Fluxes[column_name] = Anshux[column_name]
            Fluxes[error_column] = Anshux[error_column]

Fluxes.insert(0,'ID',Anshux['ID'].astype('<i4'))

Fluxes.columns = Fluxes.columns.str.replace('_err','*')
Fluxes.columns = Fluxes.columns.str.replace('_flux','')
# Fluxes[Fluxes.columns[2:]] = Fluxes[Fluxes.columns[2:]].mul(10**-9)
# # Fluxes = Fluxes.loc[:, Fluxes.columns != 'id'].mul(10**-6)
Fluxes
```

```python
Fluxes.columns = Fluxes.columns.str.replace('hst_wfc3_uvis_','')
Fluxes.columns = Fluxes.columns.str.replace('acs-wfc_','')
Fluxes.columns = Fluxes.columns.str.replace('hst_acs_','')
Fluxes.columns = Fluxes.columns.str.replace('nircam_','')
Fluxes.columns = Fluxes.columns.str.replace('wfc3-ir_','')
Fluxes.columns = Fluxes.columns.str.replace('wfc3_ir_','')
Fluxes.columns = Fluxes.columns.str.replace('jwst_','')
Fluxes.columns = Fluxes.columns.str.replace('hst_','')
Fluxes.columns = Fluxes.columns.str.replace('wfc3_uvis','')

Fluxes.columns = Fluxes.columns.str.replace('f','F')
Fluxes.columns = Fluxes.columns.str.replace('w','W')
Fluxes.columns = Fluxes.columns.str.replace('m','M')
Fluxes.columns = Fluxes.columns.str.replace('lp','LP')

Fluxes.columns = ['flux_'+ col if (idx - 1) %2 ==0 and idx>=1 else col for idx,col in enumerate(Fluxes.columns)]
Fluxes.columns = ['fluxerr_'+ col if (idx - 2) %2 ==0 and idx>=2 else col for idx,col in enumerate(Fluxes.columns)]
Fluxes.insert(len(Fluxes.columns),'spec_z',Anshux['spec_z'])
Fluxes.columns = Fluxes.columns.str.replace('*','')
# pd.set_option('max_columns', None)
# Fluxes.columns
# print(*Fluxes)

pd.set_option('display.max_columns', None)
Fluxes
```

```python
for i in range(len(Fluxes)):
    for j in range(0,20):
        if Fluxes.iloc[i,j*2]<0:
            Fluxes.iloc[i,j*2-1]=-94
            Fluxes.iloc[i,j*2]=-95
Fluxes    
```

```python
# Sample data structure representing filters and their values
filters = {
    'flux_F435W': 10,
    'fluxerr_F435W': 1,
    'flux_F555W': -5,
    'fluxerr_F555W': 0.5,
    'flux_F814W': 8,
    'fluxerr_F814W': -0.8
}

# Iterate over the filter names
for filter_name, value in Fluxes.items():
    # Check if the filter starts with 'flux_'
    if filter_name.startswith('flux_'):
        # Extract the corresponding fluxerr filter name
        fluxerr_name = 'fluxerr_' + filter_name.split('_')[1]

        # Check if the value is negative
        if value < 0:
            # Set both flux and fluxerr values to be negative
            filters[filter_name] = -abs(value)
            filters[fluxerr_name] = -abs(filters[fluxerr_name])
    if filter_name.startswith('fluxerr_'):
        # Extract the corresponding fluxerr filter name
        flux_name = 'flux_' + filter_name.split('_')[1]

        # Check if the value is negative
        if value < 0:
            # Set both flux and fluxerr values to be negative
            filters[filter_name] = -abs(value)
            filters[flux_name] = -abs(filters[flux_name])

# Print the updated filters
for filter_name, value in Fluxes.items():
    print(filter_name, ':', value)
```
