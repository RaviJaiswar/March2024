```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from astropy.io import fits
from astropy.table import Table
import os

def FancyPlot(xlab='', ylab='', lw=2.5, lfs=45, tfs=25, size_x=19, size_y=10):
    plt.rcParams['axes.linewidth'] = lw
    plt.rc('text', usetex=True)
    plt.rc('font', family='serif', size=tfs)
    mpl.rcParams['text.latex.preamble'] = [r'\usepackage{mathpazo}']
    mpl.rcParams['mathtext.default'] = 'regular'
    fig, axes = plt.subplots(1, 3, figsize=(size_x, size_y), sharex='col', sharey='row')
    axes = axes.flatten()  # Flatten the 2D array

    for axis in axes:
        axis.set_xlabel(xlab, fontsize=lfs)
        axis.set_ylabel(ylab, fontsize=lfs)
        axis.tick_params(which='major', direction='in', width=2, length=13, right=True, top=True, pad=7)
        axis.tick_params(which='minor', direction='in', width=1, length=10, right=True, top=True)
        axis.grid(False)

    # Load and plot the data in each subplot
    files = [
        ('aper_spectrum_candels-cdfs-32_132014025.fits', 4.4767),
        ('aper_spectrum_candels-cdfs-06_106007033.fits', 4.3947),
        ('aper_spectrum_candels-cdfs-06_106014046.fits', 4.6569),
#         ('aper_spectrum_candels-cdfs-22_122011067.fits', 4.2914),
#         ('aper_spectrum_candels-cdfs-22_122021111.fits', 4.7989),
#         ('aper_spectrum_candels-cdfs-23_123018120.fits', 4.5705)
    ]

    for i, (file, divisor) in enumerate(files):
        MUSE0 = Table(fits.getdata(os.path.expanduser('~/Downloads/Anshu/' + file)))
        row = i // 3
        col = i % 3
        ax = axes[i]
        FancyPlot(r'$\lambda /\AA$', r'flux/erg/s/$cm^2/\AA$', lw=2.5, lfs=45, tfs=25)
        ax.plot(MUSE0['WAVE_VAC'] / divisor, MUSE0['FLUX'])
        ax.set_xlim(1150, 1250)
        ax.set_ylim(0, 1000)  # You can adjust the ylim as needed
        plt.close(fig)  # Explicitly close the figure

    # Set labels and layout
    for ax in axes.flat:
        ax.set(xlabel=r'$\lambda /\AA$', ylabel=r'flux/erg/s/$cm^2/\AA$')
        ax.label_outer()

    plt.tight_layout()
    plt.show()

FancyPlot(r'$\lambda /\AA$', r'flux/erg/s/$cm^2/\AA$', size_x=15, size_y=20)
```
